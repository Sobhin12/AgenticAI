```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
class RState(TypedDict):
    runs: int
    fours:int
    sixes:int
    balls:int
    
    sr:float
    bpb:float
    bpercent:float
```

```python
def calc_sr(state: RState)->RState:
    sr = (state['runs']/state['balls'])*100
    state['sr'] = sr
    return state

def calc_bpb(state: RState)->RState:
    boundary = state['fours']+state['sixes']
    bpb = (state['balls']/boundary)
    state['bpb'] = bpb
    return state


def calc_bpercent(state: RState)->RState:
    boundary = state['fours']+state['sixes']
    bp = (boundary/state['runs'])
    state['bp'] = bp
    return state
```

```python
graph = StateGraph(RState)

graph.add_node('calc_bpb',calc_bpb)
graph.add_node('calc_sr',calc_sr)
graph.add_node('calc_bpercent',calc_bpercent)

graph.add_edge(START,'calc_bpb')
graph.add_edge(START,'calc_sr')
graph.add_edge(START,'calc_bpercent')
```
